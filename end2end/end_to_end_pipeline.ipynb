```python
"""
The notebook application of Text Complexity Classification for various neural network architectures in Keras.
Various accuracy and loss curves, and their accuracy and threshold scores are described in this
notebook for end to end evaluation.
"""
```

```python
# Importing libraries
import pandas as pd

from IPython.core.display import Image, display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

%matplotlib inline

# figure size in inches optional
rcParams['figure.figsize'] = 15 ,11
```

```python
def plot_images_side_by_side(image_1,image_2):
    # Takes two images and plot them side by side
    """
    image_1: image address of first image along with its extension(dtype:string)
    image_2: image address of second image along with its extension(dtype:string)
    """
    # read images
    img_A = mpimg.imread(image_1)
    img_B = mpimg.imread(image_2)

    # display images
    fig, ax = plt.subplots(1,2)
    ax[0].imshow(img_A);
    ax[1].imshow(img_B);
```
