## Build Text Dataframe

```python
import fetch_dw
text_df = fetch_dw.build_text_df(pages_df)
```

## Text Analysis by Level

```python
sns.countplot(text_df["y"])
```

```python
sns.distplot(text_df[text_df["y"] == 0]["text"].apply(lambda x: len(x.split())))
sns.distplot(text_df[text_df["y"] == 1]["text"].apply(lambda x: len(x.split())))
```

```python
sns.violinplot(y=text_df["text"].apply(lambda x: len(x.split())), x=text_df["y"])
```

```python
from textstat.textstat import textstat
text_df["FKG"] = text_df["text"].apply(textstat.flesch_kincaid_grade)

sns.distplot(text_df[text_df["y"] == 0]["FKG"], label="0")
sns.distplot(text_df[text_df["y"] == 1]["FKG"], label="1")

plt.legend()
plt.title("Distribution Flesch-Kincaid Grade per Text by Reading Lables")
plt.savefig("dist-fkg.png", dpi=300)
#files.download("dist-fkg.png")
```
